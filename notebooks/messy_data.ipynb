import pandas as pd
import numpy as np

```python
df = pd.read_csv("Desktop/Customer Churn Prediction System/data/raw/raw_telco_customer_churn.csv")
```

```python
# 1. Introduce missing values
df.loc[10:50, 'TotalCharges'] = None
df.loc[100:150, 'tenure'] = None
```

```python
# 2. Mess up categorical values
df['gender'] = df['gender'].str.lower()
df.loc[0:20, 'InternetService'] = 'dsl'
df.loc[30:60, 'PaymentMethod'] = 'credit card'
```

```python
# 3. Add noise
df.loc[200:250, 'MonthlyCharges'] = df['MonthlyCharges'] * 10
```

```python
# 4. Add duplicates
df = pd.concat([df, df.iloc[0:20]])
```

```python
# 5. Wrong data types
df['TotalCharges'] = df['TotalCharges'].astype(str)
```

```python
df.to_csv("Desktop/Customer Churn Prediction System/data/raw/raw_telco_customer_churn.csv", index=False)
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load as Series (correct structure)
top_features = pd.read_csv(
    "C:/Users/USER/Desktop/Customer Churn Prediction System/data/processed/feature_importance.csv",
    index_col=0
).iloc[:, 0]

# Sort + plot
top_features.sort_values().plot(kind='barh')

plt.title("Top Churn Drivers")
plt.xlabel("Importance Score")
plt.ylabel("Features")

plt.tight_layout()
plt.show()
```
